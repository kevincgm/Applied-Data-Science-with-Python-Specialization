```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

SPY = pd.read_csv("readonly/SPY.csv")
fund_rates = pd.read_csv("readonly/Fund_rates.csv")

# inserting missing dates when markets were not trading, flash filling with the previous trading day data
SPY['Date'] = pd.to_datetime(SPY['Date'])
SPY.set_index('Date', inplace=True)
date_range = pd.date_range(start=SPY.index.min(), end=SPY.index.max(), freq='D')
SPY = SPY.reindex(date_range)
SPY.ffill(inplace=True)

fund_rates['Date'] = pd.to_datetime(fund_rates['Date'], dayfirst=True)
fund_rates.set_index('Date', inplace=True)
fund_rates.sort_index(inplace=True)
fund_rates['Discount Rate'] = fund_rates['Discount Rate'].str.rstrip('%').str.rstrip('%[24]').astype(float)
fund_rates

# merge df and fund_rates
#df = SPY.join(fund_rates, how = 'outer')
#df = df[['Close', 'Discount Rate']]
#df
```

```python
# plotting line chart
fig, ax1 = plt.subplots(figsize = (12,6))

ax1.set_xlabel('Date', fontsize=12)
ax1.set_ylabel('SPY Closing Price', color='tab:blue', fontsize=12)
ax1.plot(SPY.index, SPY['Close'], color='tab:blue', label='SPY')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# fund rates line chart
ax2 = ax1.twinx()

ax2.set_ylabel('FED Fund Rate(%)', color='red', fontsize=12)
ax2.plot(fund_rates.index, fund_rates['Discount Rate'], color='red', label='FED Fund Rate')
ax2.tick_params(axis='y', labelcolor='red')

# covid years
start_pandemic = '2020-03-03'
start_pandemic = pd.to_datetime(start_pandemic)
plt.axvline(x=start_pandemic, color='grey', linestyle=':')

end_pandemic = '2022-03-16'
end_pandemic = pd.to_datetime(end_pandemic)
plt.axvline(x=end_pandemic, color='grey', linestyle=':', label='pandemic period')

# Add legends for both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('SPY performance vs FED Fund Rate over the past 5 years')
plt.show();
```
